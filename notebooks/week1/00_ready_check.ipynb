## 2. GPU Availability Check

```python
gpu_available = False

try:
    import cupy as cp
    print(f"CuPy version: {cp.__version__}")
    
    # Check CUDA device
    device = cp.cuda.Device(0)
    props = cp.cuda.runtime.getDeviceProperties(0)
    print(f"GPU: {props['name'].decode()}")
    print(f"Compute capability: {props['major']}.{props['minor']}")
    print(f"Total memory: {props['totalGlobalMem'] / 1e9:.1f} GB")
    
    # Quick GPU test
    x_gpu = cp.array([1, 2, 3])
    y_gpu = x_gpu * 2
    assert cp.allclose(y_gpu, cp.array([2, 4, 6]))
    print("CuPy GPU operations OK")
    gpu_available = True
    
except ImportError:
    print("CuPy not installed. Install with: pip install cupy-cuda12x")
except Exception as e:
    print(f"GPU not available: {e}")
    print("You can still follow along using Google Colab with GPU runtime.")
```

### Q1: Matrix Dimensions

If matrix A is 3x4 and matrix B is 4x2, what are the dimensions of C = A @ B?

Think about it, then run the next cell to check.

```python
# Your answer here (uncomment one):
# answer_q1 = (3, 4)
# answer_q1 = (3, 2)
# answer_q1 = (4, 4)
# answer_q1 = (4, 2)

# Verification
A = np.random.randn(3, 4)
B = np.random.randn(4, 2)
C = A @ B
print(f"A shape: {A.shape}")
print(f"B shape: {B.shape}")
print(f"C shape: {C.shape}")
print("\nRule: (M x K) @ (K x N) = (M x N)")
print("The inner dimensions (K) must match and 'cancel out'.")
```

### Q2: Linear Index

In a 5x6 matrix stored in row-major order, what linear index corresponds to element [2][3]? (0-indexed)

Formula: `linear_index = row * num_cols + col`

```python
# Your answer here:
# answer_q2 = ???

# Verification
rows, cols = 5, 6
row, col = 2, 3
linear_index = row * cols + col
print(f"Matrix shape: {rows}x{cols}")
print(f"Element [2][3] -> linear index = 2 * 6 + 3 = {linear_index}")

# Visualize
matrix = np.arange(rows * cols).reshape(rows, cols)
print(f"\nMatrix with linear indices:")
print(matrix)
print(f"\nElement at [2][3] = {matrix[2, 3]}")
```

### Q3: Floating Point Precision

How many decimal digits of precision does float32 provide?

- A) 3-4 digits
- B) 6-7 digits  
- C) 10-11 digits
- D) 15-16 digits

```python
# Verification
print("Float32 (single precision):")
print(f"  Mantissa bits: 23")
print(f"  Decimal precision: ~{np.log10(2**23):.1f} digits")
print(f"  Machine epsilon: {np.finfo(np.float32).eps}")

print("\nFloat64 (double precision):")
print(f"  Mantissa bits: 52")
print(f"  Decimal precision: ~{np.log10(2**52):.1f} digits")
print(f"  Machine epsilon: {np.finfo(np.float64).eps}")

print("\nFloat16 (half precision):")
print(f"  Mantissa bits: 10")
print(f"  Decimal precision: ~{np.log10(2**10):.1f} digits")
print(f"  Machine epsilon: {np.finfo(np.float16).eps}")
```

### Q4: Normalization

If you divide each element of [2, 4, 6] by their sum (12), what do the results sum to?

```python
# Verification
values = np.array([2, 4, 6])
total = values.sum()
normalized = values / total

print(f"Original: {values}")
print(f"Sum: {total}")
print(f"Normalized: {normalized}")
print(f"Sum of normalized: {normalized.sum()}")
print("\nThis is the basis of softmax - converting values to a probability distribution.")
```
